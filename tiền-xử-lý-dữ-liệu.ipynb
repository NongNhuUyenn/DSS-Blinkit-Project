```python
import pandas as pd
```

1. file "blinkit_orders.csv" <br>
-> tạo cột mới order_total_clean (không xoá cột cũ) = quantity*unit_price (các thuộc tính ở file "blinkit_orders_items")
2. file "blinkit_marketing_performance.csv" <br>
-> tạo cột mới roas_clean = revenue_generated/spend (không xoá cột cũ) 
3. file "blinkit_marketing_performance.csv" <br>
-> tạo thêm 1 cột target_mismatch nếu campaign = "New User Discount" và target_audience không phải "New Users".

```python
df_orders = pd.read_csv('blinkit_orders.csv')
df_order_items = pd.read_csv('blinkit_order_items.csv')
df_marketing = pd.read_csv('blinkit_marketing_performance.csv')
```

1. Order_total_clean

```python
df_order_items['order_total_clean'] = df_order_items['quantity'] * df_order_items['unit_price']
df_revenue_clean = df_order_items[['order_id', 'order_total_clean']]

df_orders_clean = pd.merge(df_orders, df_revenue_clean, on='order_id', how='inner')

df_orders_clean.to_csv('blinkit_orders_clean.csv', index=False)

print("Quy mô ma trận tệp orders mới:", df_orders_clean.shape)
df_orders_clean.head(5)
```

2. roas_clean

```python
df_marketing['roas_clean'] = df_marketing['revenue_generated'] / df_marketing['spend']

df_marketing.to_csv('blinkit_marketing_performance_clean.csv', index=False)

print("Quy mô ma trận tệp marketing mới:", df_marketing.shape)
df_marketing[['campaign_id', 'spend', 'revenue_generated', 'roas', 'roas_clean']].head(5)
```

3. target_mismatch

```python
df_marketing['target_mismatch'] = (df_marketing['campaign_name'] == 'New User Discount') & (df_marketing['target_audience'] != 'New Users')

df_marketing.to_csv('blinkit_marketing_performance_clean.csv', index=False)

print("Quy mô ma trận tệp marketing mới:", df_marketing.shape)
print("\nBảng thống kê tần suất của trường target_mismatch:")
print(df_marketing['target_mismatch'].value_counts())

df_marketing[['campaign_id', 'campaign_name', 'target_audience', 'target_mismatch']].head(5)
```
